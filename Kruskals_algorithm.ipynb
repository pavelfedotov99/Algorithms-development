```python
import numpy as np
import time
# Kruskal's algorithm in Python

class Graph:
    def __init__(self, vertices):
        self.V = vertices
        self.graph = []

    def add_edge(self, u, v, w):
        self.graph.append([u, v, w])

    # Search function

    def find(self, parent, i):
        if parent[i] == i:
            return i
        return self.find(parent, parent[i])

    def apply_union(self, parent, rank, x, y):
        xroot = self.find(parent, x)
        yroot = self.find(parent, y)
        if rank[xroot] < rank[yroot]:
            parent[xroot] = yroot
        elif rank[xroot] > rank[yroot]:
            parent[yroot] = xroot
        else:
            parent[yroot] = xroot
            rank[xroot] += 1

    #  Applying Kruskal algorithm
    def kruskal_algo(self):
        result = []
        i, e = 0, 0
        self.graph = sorted(self.graph, key=lambda item: item[2])
        parent = []
        rank = []
        for node in range(self.V):
            parent.append(node)
            rank.append(0)
        while e < self.V - 1:
            u, v, w = self.graph[i]
            i = i + 1
            x = self.find(parent, u)
            y = self.find(parent, v)
            if x != y:
                e = e + 1
                result.append([u, v, w])
                self.apply_union(parent, rank, x, y)
        for u, v, weight in result:
            print("%d - %d: %d" % (u, v, weight))
```

```python
def CreateMatrix():
    A = np.zeros(shape=(1000, 1000), dtype=int)
    count = 0
    while count < 300000:
        i = np.random.randint(0, 1000)
        j = np.random.randint(0, 1000)
        if i == j or A[i][j] != 0:
            continue
        else:
            A[i][j] = np.random.randint(1, 50)
            A[j][i] = A[i][j]
            count += 1
    return A
```

```python
A=CreateMatrix()
A
```

```python
n=1000
g = Graph(n)
for i in range(n):
    for j in range(i+1,n):
        if A[i][j]!=0:
            g.add_edge(i, j, A[i][j])
```

```python
g.kruskal_algo()
```

```python
timeitog=np.empty(n)
```

```python
for k in range(1000):
    g = Graph(k+1)
    for i in range(k+1):
        for j in range(i+1,k+1):
            if A[i][j]!=0:
                g.add_edge(i, j, A[i][j])
    start_time = time.time()
    g.kruskal_algo()
    timeitog[k]=time.time() - start_time
    print(k)
```

```python
edges = np.zeros(1000,dtype=float)
for k in range (1000):
    for i in range (k+1):
        for j in range (k+1):
            if A[i][j] != 0: 
                edges[k]+=1
    edges[k]=edges[k]/2
```

```python
from scipy.optimize import curve_fit #nlogn approximation
def func(x,a,b):
    return a*x*np.log(x)+b
popt, pcov = curve_fit(func, edges[1:1000], timeitog[1:1000])
z1=func(edges[1:1000],popt[0],popt[1])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams["figure.figsize"] = (13,13)
plt.plot(edges[1:1000],timeitog[1:1000],label ='Emphirical data')
plt.plot(edges[1:1000],z1,label ='Theory-based data')
plt.ylabel('Time (seconds)')
plt.xlabel('The number of edges')
plt.title("Kruskal's algorithm")
leg = plt.legend();
plt.show()
```

```python
from scipy.optimize import curve_fit #nlogn approximation
def func(x,a,b,c):
    return a*x*np.log(x)+b*x*x+c
popt, pcov = curve_fit(func, edges[1:1000], timeitog[1:1000])
z1=func(edges[1:1000],popt[0],popt[1],popt[2])
```
