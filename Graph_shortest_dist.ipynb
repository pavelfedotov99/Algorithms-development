```python
distances = [None] * n
parents = [None]*n
start = 1
end = 83
distances[start] = 0
queue = deque([start])

while queue:
    current_v = queue.popleft()
    for neighbour_v in AdjList[current_v]:
        if distances[neighbour_v] is None:
            distances[neighbour_v] = distances[current_v] + 1
            parents[neighbour_v] = current_v
            queue.append(neighbour_v)

path_rev = [end]
parent = parents[end]

while not parent is None:
    path_rev.append(parent)
    parent = parents[parent]

path = list(reversed(path_rev))
print("The shortest distance between node '{}' and node '{}' is {}".format(start, end, distances[end]) )

if path != [end]:
    print('The shortest path is {}'.format(path))
```
